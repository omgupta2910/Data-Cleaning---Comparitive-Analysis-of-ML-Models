#Clustering

```python
from sklearn.cluster import KMeans
```

```python
x = "perimeter"
y = "asymetry"
X = df[[x,y]].values
```

```python
kmeans = KMeans(n_clusters=3).fit(X)
```

```python
cluster = kmeans.labels_
```

```python
cluster_df = pd.DataFrame(np.hstack((X,cluster.reshape(-1,1))),columns = [x,y,"class"])
```

```python
# K_means classes
sns.scatterplot(x=x, y=y , hue='class', data=cluster_df)
plt.plot()
```

```python
# Original Classes
sns.scatterplot(x=x, y=y , hue='class', data=df)
plt.plot()
```
